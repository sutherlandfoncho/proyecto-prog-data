```python
import sys
import os
import pandas as pd
# Importamos tus funciones desde el script que creamos
sys.path.append('../src')
from validation import calcular_checksum, validar_reglas_negocio, check_calidad_basica

# Ruta al archivo que generó Diego
ruta_limpio = '../data/processed/route_cost_impact_limpio.csv'

if os.path.exists(ruta_limpio):
    md5_check = calcular_checksum(ruta_limpio)
    print(f"--- VALIDACIÓN DE PROCEDENCIA ---")
    print(f"Archivo: {os.path.basename(ruta_limpio)}")
    print(f"MD5 Hash: {md5_check}")
    print(f"Estado: Verificado")
else:
    print("ERROR: No se encuentra el archivo limpio en la ruta especificada.")
```

```python
# Cargamos el dataset para las pruebas
df = pd.read_csv(ruta_limpio)

# 1. Ejecutar reglas de negocio del script src/validation.py
test_resultados = validar_reglas_negocio(df)

# 2. Crear tabla comparativa de filas
# antes de dropping
data_comparativa = {
    "Etapa": ["Original (Post-Limpieza)", "Validación Actual"],
    "Total Filas": [3132, len(df)],
    "Estado": ["OK", "Coincide" if len(df) == 3132 else "Discrepancia"]
}

df_comparativo = pd.DataFrame(data_comparativa)

print("--- RESULTADOS DE VALIDACIÓN TÉCNICA ---")
print("\n1. Consistencia de Filas:")
display(df_comparativo)

print("\n2. Pruebas de Reglas de Negocio:")
for prueba, pasado in test_resultados.items():
    status = "✅ PASÓ" if pasado else "❌ FALLÓ"
    print(f"- {prueba.replace('_', ' ').title()}: {status}")
```
